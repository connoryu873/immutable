```python
df = cleaned_data.copy()
```

```python
#Assign alpha values for EWMQ for each quality type
alpha_dict = {'meteorite': 0.3, 'diamond': 0.05, 'gold': 0.1, 'shadow': 0.15}

# Give date range
date_range = pd.date_range(start='2021-12-15', end='2023-05-15', freq='D')


# Create an empty DataFrame with the date range as the index
df_ewma = pd.DataFrame(index=date_range)
df_asset_quality = df[['asset_name','asset_quality']].drop_duplicates()

# Apply EWMA
for index, row in tqdm_notebook(df_asset_quality.iterrows(), desc="Generating EWMA prices"):
    
    asset_name = row['asset_name']
    asset_quality = row['asset_quality']

    filt_asset_name = df['asset_name'] == asset_name
    filt_asset_quality = df['asset_quality'] == asset_quality
    #Combine each card and asset quality 
    temp_df = df[filt_asset_name & filt_asset_quality]
    alpha = alpha_dict[asset_quality]

    #Apply EWMA model onto particular card and asset
    ewma_data = temp_df['average_sale_price'].ewm(alpha = alpha,halflife = '150 days' ,adjust=True, times = temp_df['snapshot_date'] ).mean()

    #Temporarily assign card and asset
    temp_series = pd.DataFrame({f'{asset_name}+{asset_quality}': ewma_data.values, },
    index= temp_df['snapshot_date'])
    
    #Merge onto main df
    df_ewma = df_ewma.merge(temp_series, left_index=True, right_index=True, how='left')

    
    

df_ewma.ffill(inplace=True) #forward fill days where prices do not exist
    
```

# Generating Circulation Supply dataframe where
- index = snapshot dates
- column names is asset name + asset quality
- cells represent circulation supply and nulls are forward filled if possible

```python
df_ewma
```

```python
#Create df of circulation of all card + quality 
df_circulation = df.copy(deep=True).set_index('snapshot_date')
df_circulation["asset_name_quality"] = df_circulation["asset_name"].astype(str) + '+' + df_circulation["asset_quality"].astype(str)


df_circulation = df_circulation.pivot(columns=['asset_name_quality'], values='assets_in_circulation')
df_circulation
df_circulation.ffill(inplace=True)
df_circulation
```

```python
# Align rows and columns based on matching row indices and column names
matching_rows = df_circulation.index.intersection(ewma.index)
matching_columns = df_circulation.columns.intersection(ewma.columns)

df1_aligned = df_circulation.loc[matching_rows, matching_columns]
df2_aligned = ewma.loc[matching_rows, matching_columns]

# Perform matrix multiplication
result = df1_aligned * df2_aligned

# resulting df
result
```

```python
# Create row sum for each result
row_sum = result.sum(axis=1)

# Add the row-wise sum as a new column as mcap
result['market_cap'] = row_sum

result['market_cap'] = result['market_cap'].apply(lambda x: "{:.2f}".format(x))

```

```python
#Create chart to map EWMA mcap vs date
results_plot = result.copy()
results_plot = results_plot['market_cap']

results_plot = pd.DataFrame(results_plot)
results_plot.reset_index(inplace=True)

results_plot['market_cap'] = results_plot['market_cap'].astype(float)

fig = px.line(results_plot, x='snapshot_date', y='market_cap', title='EWMA market cap over time')
fig.show()
```
